```python
def plot_sub(row,col,xarr,yarr,title,ax_arr,plot_opts):
    ax_arr[row, col].plot(xarr,yarr,plot_opts)
    ax_arr[row, col].text(0.5, 1.08, title,
         horizontalalignment='center',
         fontsize=20,
         transform = ax_arr[row,col].transAxes)
    ax_arr[row,col].set_xlim([np.min(xarr), np.max(xarr)])
    
    
#plot grid parameters
Lval=6.0
Lidx =  min(np.where(r_arr[:,num_u3/2]>=Lval)[0])
pylab.rcParams['figure.figsize'] = (20, 12)

# Four axes, returned as a 2-d array
f, axarr = plt.subplots(3, 4)

plot_sub(0,0,u3,gsup11[Lidx,:],r'$g_{11_{con}}$',axarr,'-b')
plot_sub(0,1,u3,gsup22[Lidx,:],r'$g_{22_{con}}$',axarr,'-b')
plot_sub(0,2,u3,gsup33[Lidx,:]*dmudx[Lidx,:]**2,r'$g_{33_{con}}$',axarr,'-b')
plot_sub(0,3,u3,gsup13[Lidx,:]*dmudx[Lidx,:],r'$g_{13_{con}}$',axarr,'-b')

plot_sub(1,0,u3,g11[Lidx,:],r'$g_{11_{cov}}$',axarr,'-b')
plot_sub(1,1,u3,g22[Lidx,:],r'$g_{22_{cov}}$',axarr,'-b')
plot_sub(1,2,u3,g33[Lidx,:]/dmudx[Lidx,:]**2,r'$g_{33_{cov}}$',axarr,'-b')
plot_sub(1,3,u3,g13[Lidx,:]/dmudx[Lidx,:],r'$g_{13_{cov}}$',axarr,'-b')

plot_sub(2,0,u3,Jac[Lidx,:]/dmudx[Lidx,:],r'$Jacobian$',axarr,'-b')
plot_sub(2,1,u3,r_arr[Lidx,:],r'$R_{arr}$',axarr,'-b')
plot_sub(2,2,u3,arccos(costh[Lidx,:])*180/pi,r'$\theta_{arr}$',axarr,'-b')

axarr[2, 3].plot(x_arr,y_arr,'og',ms=0.3)
axarr[2, 3].text(  0.5, 1.08, r'$Grid$',
                        horizontalalignment='center',
                        fontsize=20,
                        transform = axarr[2,3].transAxes)
axarr[2,3].set_xlim([0, 10])

axarr[2, 3].plot(x_arr[Lidx,:],y_arr[Lidx,:],'-r')
axarr[2, 3].text(  0.5, 1.08, r'$Grid$',
                        horizontalalignment='center',
                        fontsize=20,
                        transform = axarr[2,3].transAxes)
    
subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95, wspace=0.25, hspace=0.25)
f.savefig('../fortran/plots/test.pdf',format='PDF')
```
