Open up the plotting constants file. This contains:

- num_u1
- num_u3
- va(num_u1,num_u3)
- x_arr(num_u1,num_u3)
- y_arr(num_u1,num_u3)
- Eidx(num_u1/2+1)
- dx2

```python
%pylab inline
filename = '../fortran/data/plotting_constants.dat'
f = open ( filename , 'rb' )
```

```python
#Cordinate Definitions
num_u1 = np.fromfile ( f , dtype = np.int32 , count = 1 )[0]
num_u3 = np.fromfile ( f , dtype = np.int32 , count = 1 )[0]
#Va Profile
va = np.fromfile ( f , dtype = np.float64 , count = num_u1*num_u3 )
#Grid Arrays
x_arr = np.fromfile ( f , dtype = np.float64 , count = num_u1*num_u3 )
y_arr = np.fromfile ( f , dtype = np.float64 , count = num_u1*num_u3 )
#other
Eidx = np.fromfile ( f , dtype = np.int32 , count = num_u1/2+1 )
dx2 = np.fromfile ( f , dtype = np.float64 , count = 1 )[0]
```

```python
f.close()
```

```python
#Reshape arrays to 2D
#'F' means they're fortran refernced(Column Major)
va = np.reshape(va,(num_u1,num_u3),order='F')
x_arr = np.reshape(x_arr,(num_u1,num_u3),order='F')
y_arr = np.reshape(y_arr,(num_u1,num_u3),order='F')
```

```python
print 'Num_u1:',num_u1
print 'Num_u3:',num_u3
```

```python
#define default figure size
pylab.rcParams['figure.figsize'] = (20, 16)

#plot the grid
plot(x_arr,y_arr,'ro',ms=4);
```
