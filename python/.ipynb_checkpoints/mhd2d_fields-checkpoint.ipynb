Open up the plotting constants file. This contains:

- num_u1
- num_u3
- va(num_u1,num_u3)
- x_arr(num_u1,num_u3)
- y_arr(num_u1,num_u3)
- Eidx(num_u1/2+1)
- dx2

```python
#Open plotting constants
filename = '../fortran/data/plotting_constants.dat'
f = open ( filename , 'rb' )

#Cordinate Definitions
num_u1 = np.fromfile (f,dtype = np.int32,count = 1)[0]
num_u3 = np.fromfile (f,dtype = np.int32,count = 1)[0]

#Va Profile
va = np.fromfile (f,dtype = np.float64,count = num_u1*num_u3)

#Grid Arrays
x_arr = np.fromfile (f,dtype = np.float64,count = num_u1*num_u3)
y_arr = np.fromfile (f,dtype = np.float64,count = num_u1*num_u3)

#other
Eidx = np.fromfile (f,dtype = np.int32,count = num_u1/2+1)
dx2 = np.fromfile (f,dtype = np.float64,count = 1)[0]

#close file
f.close()

#Reshape arrays to 2D
#'F' means they're fortran refernced(Column Major)
va = np.reshape(va,(num_u1,num_u3),order='F')
x_arr = np.reshape(x_arr,(num_u1,num_u3),order='F')
y_arr = np.reshape(y_arr,(num_u1,num_u3),order='F')
```

Open up the data plotting files. These contain:

Time Counter:

- tt: double precision time counter

Fields:

- E_nu: dcmplex (num_u1,num_u3)
- E_ph: dcmplex (num_u1,num_u3)
- E_mu: dcmplex (num_u1,num_u3)
- B_nu: dcmplex (num_u1,num_u3)
- B_ph: dcmplex (num_u1,num_u3)
- B_mu: dcmplex (num_u1,num_u3)

Ionosphere Values:

- b1_n,b1atm_n: dcmplex (num_u1)
- b2_n,b2atm_n: dcmplex (num_u1)
- b1_s,b1atm_s: dcmplex (num_u1)
- b2_s,b2atm_s: dcmplex (num_u1)

Psi:

- psiatm_N: dcmplex (num_u1)
- psiatm_S: dcmplex (num_u1)

I imaginary number:

- im: dcmplex (0,m)

bsup3:

- bsup3_n: dcmplex (num_u1,num_u3)
- bsup3_s: dcmplex (num_u1,num_u3)

hr scaling:

- hratm_n: double precision (num_u1)
- hratm_s: double precision (num_u1)

Ground Fields:

- bxg_n: dcmplex (num_u1)
- byg_n: dcmplex (num_u1)
- bxg_s: dcmplex (num_u1)
- byg_s: dcmplex (num_u1)

```python
plot_start=1
plot_finish=10
```

```python
#Open fields
ii=100
ii_str = '%04d' % ii
filename = '../fortran/data/plot_data'+ ii_str +'.dat'
f = open ( filename , 'rb' )

tt =  np.fromfile (f,dtype = np.float64,count = 1)[0]
e_nu = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))
e_ph = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))
e_mu = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))

b_nu = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))
b_ph = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))
b_mu = (np.fromfile (f,dtype = np.complex128,count = num_u1*num_u3))

#close file
f.close()

#Reshape arrays to 2D
#'F' means they're fortran refernced(Column Major)
e_nu = np.reshape(e_nu,(num_u1,num_u3),order='F')
e_ph = np.reshape(e_ph,(num_u1,num_u3),order='F')
e_mu = np.reshape(e_mu,(num_u1,num_u3),order='F')

b_nu = np.reshape(b_nu,(num_u1,num_u3),order='F')
b_ph = np.reshape(b_ph,(num_u1,num_u3),order='F')
b_mu = np.reshape(b_mu,(num_u1,num_u3),order='F')

pylab.rcParams['figure.figsize'] = (20,12)
phi=22.5
im = 0+2.0j

print 'Max of E_nu:',abs(e_nu).max()
print 'At',tt, 'seconds'

fig = figure()
fig.patch.set_facecolor('white')
ax = fig.add_subplot(111)
pcolormesh(x_arr[0:num_u1-1:2,0:num_u3-1:2],\
           y_arr[0:num_u1-1:2,0:num_u3-1:2],\
           e_nu[0:num_u1-1:2,0:num_u3-1:2]*exp(im*phi*pi/180),\
           shading='gouraud',\
           cmap=cm.rainbow);
ax.set_title(r'$E_{\nu}$',fontsize=25)
ax.patch.set_facecolor('black')
colorbar()
```

```python
# Four axes, returned as a 2-d array
f, axarr = subplots(2, 3)
### e_nu
axarr[0,0].pcolormesh(    x_arr[0:num_u1-1:2,0:num_u3-1:2],\
                          y_arr[0:num_u1-1:2,0:num_u3-1:2],\
                          e_nu[0:num_u1-1:2,0:num_u3-1:2]*exp(im*phi*pi/180),\
                          shading='gouraud',\
                          cmap=cm.rainbow);

axarr[0,0].patch.set_facecolor('black')
axarr[0, 0].set_title(r'$E_{\nu}$',fontsize=20)
### e_ph
axarr[0,1].pcolormesh( x_arr[1:num_u1-1:2,0:num_u3-1:2],\
                       y_arr[1:num_u1-1:2,0:num_u3-1:2],\
                       e_ph[1:num_u1-1:2,0:num_u3-1:2]*exp(im*phi*pi/180),\
                       shading='gouraud',\
                       cmap=cm.rainbow);

axarr[0,1].patch.set_facecolor('black')
axarr[0, 1].set_title(r'$E_{\phi}$',fontsize=20)
### e_mu
axarr[0,2].pcolormesh(  x_arr[1:num_u1-1:2,1:num_u3-1:2],\
                        y_arr[1:num_u1-1:2,1:num_u3-1:2],\
                        e_mu[1:num_u1-1:2,1:num_u3-1:2]*exp(im*phi*pi/180),\
                        shading='gouraud',\
                        cmap=cm.rainbow);

axarr[0,2].patch.set_facecolor('black')
axarr[0, 2].set_title(r'$E_{\mu}$',fontsize=20)
### b_nu
axarr[1,0].pcolormesh( x_arr[1:num_u1-1:2,1:num_u3-1:2],\
                       y_arr[1:num_u1-1:2,1:num_u3-1:2],\
                       b_nu[1:num_u1-1:2,1:num_u3-1:2]*exp(im*phi*pi/180),\
                       shading='gouraud',\
                       cmap=cm.rainbow);

axarr[1,0].patch.set_facecolor('black')
axarr[1,0].set_title(r'$B_{\nu}$',fontsize=20)
### b_ph
axarr[1,1].pcolormesh( x_arr[0:num_u1-1:2,1:num_u3-1:2],\
                       y_arr[0:num_u1-1:2,1:num_u3-1:2],\
                       b_ph[0:num_u1-1:2,1:num_u3-1:2]*exp(im*phi*pi/180),\
                       shading='gouraud',\
                       cmap=cm.rainbow);

axarr[1,1].patch.set_facecolor('black')
axarr[1,1].set_title(r'$B_{\phi}$',fontsize=20)
### b_mu
axarr[1,2].pcolormesh( x_arr[0:num_u1-1:2,0:num_u3-1:2],\
                       y_arr[0:num_u1-1:2,0:num_u3-1:2],\
                       b_mu[0:num_u1-1:2,0:num_u3-1:2]*exp(im*phi*pi/180),\
                       shading='gouraud',\
                       cmap=cm.rainbow);
axarr[1,2].set_title(r'$B_{\mu}$',fontsize=20)
axarr[1,2].patch.set_facecolor('black')
```
